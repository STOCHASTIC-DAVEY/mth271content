**Task 1:** Consider a circle divided into $n$ equal arcs, each of which is colored in either red or blue. In each iteration,  one of the arcs is chosen at random. The selected arc then chooses one of its neighbors at random and adopts that neighbor's color.  In the case $n=3$, when does the iteration stop changing the colors? In the case $n=3$, does the final state depend on the initial color configuration? 

**Task 2:** Model the process in Task 1 for a general $n$ as a Markov chain whose states consist of all of the possible color configurations on the arcs. How many states are there?  What are the absorbing states?  Is this an absorbing Markov chain? How would you compute the probabilities of hitting the absorbing states in any number of iterations?

**Task 3:** Generalize from a circle to the surface of a torus. The toroidal  surface is divided into $n \times n$ rectangular patches each of which can adopt  one of $k$ colors (see the figure below for an example). Now each rectangle has 8 neighbors from whom to adopt a color in each iteration. Answer the questions of Task 2 in this situation. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(); ax = fig.gca(projection='3d')
angs = np.linspace(0, 2.*np.pi, 20)
theta, phi = np.meshgrid(angs, angs)
x = (2 + np.cos(theta)) * np.cos(phi)
y = (2 + np.cos(theta)) * np.sin(phi)
z = np.sin(theta)
rng = np.random.default_rng()
randind = rng.integers(5, size=x.shape)
colors = np.array(['crimson', 'coral', 'gold', 'blue', 'olive'])[randind]
ax.plot_surface(x, y, z, facecolors=colors, linewidth=1, edgecolors='k')
ax.view_init(46, 26); ax.set_zlim(-3,3);
```
